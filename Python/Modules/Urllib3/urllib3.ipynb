## fetching url details

```python
from urllib.parse import urlparse, urlunparse
parse_url = urlparse('https://baivabsarkar.netlify.app/')
print(parse_url)
print("\n")
unparse_url = urlunparse(parse_url)
print(unparse_url)
```

## fetching HTML

```python
from urllib.request import urlopen
with urlopen('https://baivabsarkar.netlify.app/') as response:
     for line in response:
         line = line.decode('utf-8')
         print(line)
```

## fetching HTML and write the bytes directly to a file without having to decode

```python
from urllib.request import urlopen
import json
url = "https://jsonplaceholder.typicode.com/todos/1"
with urlopen(url) as response:
    body = response.read()

todo_item = json.loads(body)
todo_item
```

## locally save an image using Urllib

```python
import urllib.request
import os

opener=urllib.request.build_opener()
opener.addheaders=[('User-Agent','Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36')]
urllib.request.install_opener(opener)

# Image URL and Filename
filename = "pic.jpg"
image_url = "https://cdn.pixabay.com/photo/2016/11/19/14/00/code-1839406_960_720.jpg"
urllib.request.urlretrieve(image_url, filename)
print(os.listdir())
```

## locally save a pdf using Urllib

```python
import requests

URL= "https://www.clickdimensions.com/links/TestPDFfile.pdf"

response = requests.get(URL, stream = True)

with open("large-file.pdf","wb") as pdf:
    for chunk in response.iter_content(chunk_size=1024):
        if chunk:
            pdf.write(chunk)
```

```python
from urllib.request import urlopen
with urlopen("https://www.example.com") as response:
    body = response.read()

with open("example.html", mode="wb") as html_file:
    html_file.write(body)
```

## urllib.request Represents an HTTP Message

```python
from urllib.request import urlopen
from pprint import pprint
with urlopen("https://www.example.com") as response:
    pprint(dir(response))
```
