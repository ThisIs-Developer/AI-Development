```python
import pandas as pd
df=pd.read_csv("Mall_Customers.csv")
print(df)
x=df.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import KMeans
wess=[]
for i in range(1,26):
  km=KMeans(n_clusters=i,init='k-means++',random_state=0)
  km.fit(x)
  inr=km.inertia_
  wess.append(inr)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,26),wess)
plt.xlabel("No of cluster")
plt.ylabel("itertia")
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=5,init='k-means++',random_state=1)
y_predict=km.fit_predict(x)
print(y_predict)
```

```python
print(len(x))
print(len(y_predict))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=10,c='red',label="cluster0")
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=10,c='blue',label="cluster1")
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],s=10,c='green',label="cluster2")
plt.scatter(x[y_predict==3,0],x[y_predict==3,1],s=10,c='blueviolet',label="cluster3")
plt.scatter(x[y_predict==4,0],x[y_predict==4,1],s=10,c='hotpink',label="cluster4")
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[y_predict==0,0],x[y_predict==0,1],s=10,c='red',label="cluster0")
plt.scatter(x[y_predict==1,0],x[y_predict==1,1],s=10,c='blue',label="cluster1")
plt.scatter(x[y_predict==2,0],x[y_predict==2,1],s=10,c='green',label="cluster2")
plt.scatter(x[y_predict==3,0],x[y_predict==3,1],s=10,c='blueviolet',label="cluster3")
plt.scatter(x[y_predict==4,0],x[y_predict==4,1],s=10,c='hotpink',label="cluster4")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=130,marker='^',c='black')
plt.xlabel("Income")
plt.ylabel("Score")
plt.title("Mall Customers")
plt.show()
```
