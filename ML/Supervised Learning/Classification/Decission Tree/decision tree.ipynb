## DecisionTreeClassifier
**DecisionTreeClassifier is a classification algorithm in scikit-learn that uses a tree-like model to make predictions by recursively splitting the data based on feature conditions. It's a popular and interpretable algorithm for classification tasks.**

```python
import pandas as pd
df = pd.read_csv("Social_Network_Ads.csv")
x = df.iloc[:, [2, 3]].values
y = df.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_predict = dt.predict(x_test)
accuracy = accuracy_score(y_test, y_predict)
print(accuracy)
```
