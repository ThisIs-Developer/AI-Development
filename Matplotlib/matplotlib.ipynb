```python
import matplotlib.pyplot as plt
year=[1960,1970,1980,1990,2000,2010]
pop_aus=[44.91,58.09,78.07,107.7,138.5,170.6]
pop_ind=[149.48,559.57,696.783,270.133,1000.4,309.1]

plt.plot(year, pop_aus, color='green')
plt.plot(year, pop_ind, color='red')

plt.grid(True)
plt.xlabel("Year(S)")
plt.ylabel("Population in million")
plt.title("Australia vs India Popultaion")
plt.show()
```

```python
from pylab import *
t = arange(0.0, 2.0, 0.01)
s = 1 + sin(2.2*pi*t)
plt.plot(t, s)

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Sin Wave')
plt.grid(True)
savefig("Sin_wave.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-10,9,20)
y=x**3
plt.subplot(2,2,1)
plt.plot(x,y,"b*-")
plt.subplot(2,2,2)
plt.plot(x,y,"gP-")
plt.subplot(2,2,3)
plt.plot(x,y,"ms-")
plt.subplot(2,2,4)
plt.plot(x,y,"r^-")
```
