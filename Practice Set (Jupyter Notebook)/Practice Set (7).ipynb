```python
def frequency(word):
  dict={}
  for i in word:
    keys=dict.keys()
    if i in keys:
      dict[i]+=1
    else:
      dict[i]=1
  return dict
word=str(input("Enter a word = "))
print(frequency(word))
```

```python
# baivab
# ba  ab
# baab
def add_string(word):
  if(len(word)<2):
    return ' '
  else:
    return word[0:2]+word[-2:]
word=str(input("Enter a word = "))
print(add_string(word))
```

```python
def replace_string(word):
  temp=word[0]
  word=word.replace(temp, '$')
  word=temp+word[1:] #logic
  return word
word=str(input("Enter a word = "))
print(replace_string(word))
```

```python
# abc
# xyz
# xyab
def one_string(word1,word2):
  newword1= word2[0:2] + word1[2:]
  newword2= word1[0:2] + word2[2:]
  return newword1 + ' ' + newword2
word1=str(input("Enter a word = "))
word2=str(input("Enter a word = "))
print(one_string(word1,word2))
```
