6. Write a Python program to get the Fibonacci series between 0 to 50.

```python
def fibonacci(n):
  if n==1 or n==0:
    return n
  else:
    return fibonacci(n-1)+fibonacci(n-2)
n=int(input("Enter the Length of the Series : "))
for i in range(0,n):
  print(fibonacci(i))
```

7. Write a Python program which iterates the integers from 1 to 50. For multiples
of three print &quot;Fizz&quot; instead of the number and for the multiples of five print
&quot;Buzz&quot;. For numbers which are multiples of both three and five print &quot;FizzBuzz&quot;.

```python
for num in range (1,51):
  if num%15==0:
    print('FIZZBUZZ')
  elif num%5==0:
    print('BUZZ')
  elif num%3==0:
    print('FIZZ')
  else:
    print(num)
```

8. Write a Python program which accepts a sequence of comma separated 4-digit
binary numbers as its input and print the numbers that are divisible by 5 in a
comma separated sequence.

```python
items = []
num = [x for x in input().split(',')]
for p in num:
    x = int(p, 2)
    if not x%5:
        items.append(p)
print(','.join(items))
```

11. Write a Python program to calculate a dog&#39;s age in dog&#39;s years. For the first two
years, a dog year is equal to 10.5 human years. After that, each dog year equals 4
human years.

```python
age=int(input("Enter the age of the Dog : "))
if(age>0):
    if(age<=2):
        print("Dog's age is :",age,"Which is equal to",age*10.5,"Human age")
    else:
        print("Dog's age is :",age,"Which is equal to",(2*10.5)+(age-2)*4,"Human age")
else:
    print("Dog's is not yet birth")
```

12. Write a Python program to check whether an alphabet is a vowel or consonant.

```python
alpha=input("Enter Alphabet : ")
if(alpha=='A' or alpha=='E' or alpha=='I' or alpha=='O' or  alpha=='U'):
    print("Alphabet is Vowel")
elif(alpha=='a' or alpha=='e' or alpha=='i' or alpha=='o' or  alpha=='u'):
    print("Alphabet is Vowel")
else:
    print("Alphabet is consonant")
```
