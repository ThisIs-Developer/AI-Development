```python
def factorial(num):
  fact=1
  for i in range(1,num+1):
    fact=fact*i;
  print("Factorial : ",fact)
num=int(input("Entre the number : "))
factorial(num)
```

```python
def cal(a,b):
  next=1
  while(next!=0):
    a=int(input("Enter the first number : "))
    b=int(input("Enter the second number : "))
    print("Operation : + - * /")
    choice=input("Enter your choice : ")
    if(choice == '+'):
      print("Sum : ",a+b)
    elif(choice == '-'):
      print("Subtraction : ",a-b)
    elif(choice == '*'):
      print("Multipication : ",a*b)
    elif(choice == '/'):
      print("Division : ",a/b)
    else:
      print("Enter choice correctly")
    print("Do you want to continue")
    next=int(input())
cal(a,b)
```

```python
def fibo(num):
  if(num==0):
    return num
  elif(num==1):
    return num
  else:
    return fibo(num-1)+fibo(num-2)
num=int(input("Enter the nmber : "))

for i in range(0,num):
  print(fibo(i))
```
