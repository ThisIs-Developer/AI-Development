```python
import numpy as np
x=[2,4,6,8,10]
w=[0.01,0.03,0.05,0.007,0.009]
b0=0
wT=np.transpose(w)
Y=0
for i in range(len(x)):
  Y=Y+wT[i]*x[i]
Y=b0+Y
if Y>=1:
  print("Y=1")
else:
  print("Y=0")
```

```python
import numpy as np
x=np.array(([1,1,0],[1,0,0],[1,0,1],[1,1,1] ))
w=[0.01,0.03,0.005]
b0=0
wT=np.transpose(w)
Y=0
for j in range(len(w)):
  for i in range(len(x[1])):
    Y=Y+wT[i]*x[i][j]
Y=b0+Y
if Y>=1:
  print("Y=1")
else:
  print("Y=0")
```
